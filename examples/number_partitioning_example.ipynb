```python
num_set = [8, 7, 6, 5, 4]
```

```python
number_partitioning = NumberPartitioning(num_set)
```

```python
qp = number_partitioning.to_quadratic_program()
qp
```

```python
optimizer = CplexOptimizer()
result = optimizer.solve(qp)
result
```

```python
print(number_partitioning.interpret(result))
```
