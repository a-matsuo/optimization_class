```python
%load_ext autoreload
%autoreload 2

import networkx as nx

from qiskit_optimization.applications.ising.set_packing import SetPacking
from qiskit_optimization.algorithms import CplexOptimizer
```

```python
U = [1, 2, 3, 4, 5]
list_of_subsets = [[1, 2, 3], [2, 3, 4], [4, 5], [1, 3], [2]]
```

```python
set_packing = SetPacking(list_of_subsets)
```

```python
qp = set_packing.to_quadratic_program()
qp
```

```python
optimizer = CplexOptimizer()
result = optimizer.solve(qp)
result
```

```python
print(set_packing.interpret(result))
```
