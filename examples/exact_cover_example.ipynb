```python
%load_ext autoreload
%autoreload 2

import networkx as nx

from qiskit_optimization.applications.exact_cover import ExactCover
from qiskit_optimization.algorithms import CplexOptimizer
```

```python
U = [1, 2, 3, 4, 5, 6, 7]
list_of_subsets = [[1, 4, 7], [1, 4], [4, 5, 7], [3, 5, 6], [2, 3, 6, 7], [2, 7]]
```

```python
exact_cover = ExactCover(list_of_subsets)
```

```python
qp = exact_cover.to_quadratic_program()
qp
```

```python
optimizer = CplexOptimizer()
result = optimizer.solve(qp)
result
```

```python
print(exact_cover.interpret(result))
```
