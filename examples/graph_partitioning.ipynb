```python
%load_ext autoreload
%autoreload 2

import networkx as nx

from qiskit_optimization.applications.ising.graph_partitioning import GraphPartitioning
from qiskit_optimization.applications.ising.max_cut import Maxcut

from qiskit_optimization.algorithms import MinimumEigenOptimizer, RecursiveMinimumEigenOptimizer, CplexOptimizer
```

```python
g=nx.gnm_random_graph(10,15, 123)
pos = nx.spring_layout(g, seed=111)
nx.draw(g, pos=pos, with_labels=True)
```

```python
graph_partitioning = GraphPartitioning(g)
```

```python
qp = graph_partitioning.to_quadratic_program()
qp
```

```python
optimizer = CplexOptimizer()
result = optimizer.solve(qp)
result
```

```python
print(graph_partitioning.interpret(result.x))
graph_partitioning.plot_graph(result.x, pos=pos)
```
